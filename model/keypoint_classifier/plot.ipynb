```python
import time
def plotgrapgh(r1,r2,name):

    with open(name, 'r') as file_obj:
        csv_reader=reader(file_obj)
        dataset=list(csv_reader)
        for items in dataset:
            for item in items:
                i=items.index(item)
                temp=item.capitalize();
                items[i]=temp
    a=0
    point_x=[]
    point_y=[]
    for data in dataset[0][1:]:
        if a%2 ==0:
            point_x.append(float(data))
        else:
            point_y.append(float(data))
        a += 1

    print(point_x)

    import matplotlib.pyplot as plt

    # plt.scatter(point_x,point_y)


    for m in range(r1,r2):

        dpoint=dataset[m][1:]

        a=0
        point_x=[]
        point_y=[]
        for data in dpoint:
            if a%2 ==0:
                point_x.append(float(data))
            else:
                point_y.append(float(data))
            a += 1

        xline1=list()
        xline2=list()
        xline3=list()
        xline4=list()
        xline5=list()
        xline6=list()
        yline1=list()
        yline2=list()
        yline3=list()
        yline4=list()
        yline5=list()
        yline6=list()
        for i in range(0,5):
            xline1.append(point_x[i])
            yline1.append(point_y[i])

        xline2.append(point_x[0])
        yline2.append(point_y[0])
        for i in range(5,9):
            xline2.append(point_x[i])
            yline2.append(point_y[i])


        for i in range(9,13):
            xline3.append(point_x[i])
            yline3.append(point_y[i])

        for i in range(13,17):
            xline4.append(point_x[i])
            yline4.append(point_y[i])


        xline5.append(point_x[0])
        yline5.append(point_y[0])

        for i in range(17,21):
            xline5.append(point_x[i])
            yline5.append(point_y[i])

        xline6.append(point_x[5])
        xline6.append(point_x[9])
        xline6.append(point_x[13])
        xline6.append(point_x[17])


        yline6.append(point_y[5])
        yline6.append(point_y[9])
        yline6.append(point_y[13])
        yline6.append(point_y[17])



        import matplotlib.pyplot as plt

        plt.plot(xline1,yline1)
        plt.plot(xline2,yline2)
        plt.plot(xline3,yline3)
        plt.plot(xline4,yline4)
        plt.plot(xline5,yline5)
        plt.plot(xline6,yline6)
        plt.xlabel('fig'+str(m))
```
